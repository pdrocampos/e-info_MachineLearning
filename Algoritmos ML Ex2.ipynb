```python
from sklearn.datasets import load_iris

import pandas as pd

# Coleta dos dados
iris_data = load_iris(as_frame=True)

df = pd.DataFrame(iris_data['data'])
df['label'] = iris_data['target'].astype(str)
df['label'] = ['setosa' if t == '0' else t for t in df['label']]
df['label'] = ['versicolor' if t == '1' else t for t in df['label']]
df['label'] = ['virginica' if t == '2' else t for t in df['label']]
```

```python
from sklearn.model_selection import train_test_split

data_train, data_test = train_test_split(df, test_size = 0.3)

X_train = data_train[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
y_train = data_train['label'].values

X_test = data_test[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
y_test = data_test['label'].values
```

```python
from sklearn.cluster import KMeans

ml_model = KMeans(n_clusters=3)
ml_model = ml_model.fit(X_train)
```

```python
y_pred = ml_model.predict(X_test)

df_test = pd.DataFrame(X_test)
df_test['cluster'] = y_pred
df_test['label'] = y_test
df_test
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data = df_test, x = 'petal length (cm)', y = 'petal width (cm)', hue = 'cluster', palette = 'Set1');
plt.show();
```
